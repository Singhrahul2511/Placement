```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

Basics graph.

```python
x= [5,10,15,20,25]
y =[6,12,18,24,30]
plt.plot(x,y)
plt.show()#printing the graph without showing the values
```

```python
#you can change the color of line
plt.plot(x,y,color = 'red')
```

```python
#for marking the points.
plt.plot(x,y,color ='red',marker = '*')
plt.show()
```

```python
#you can change the color of marker also.
plt.plot(x,y,color = 'red',marker ='*',markeredgecolor = 'black',markersize = '10')
```

```python
#you can also give iit as a dashdot.
plt.plot(x,y,color = 'red',marker ='*',linestyle ='--',markeredgecolor='black',markersize='10')
```

```python
#short form of (color,marker and linestyle)
plt.plot(x,y,'b*--')
```

```python
#for heading of the graph.
plt.plot(x,y,'b*--')
plt.title('Heading1',fontdict={'fontname': 'Comic Sans Ms','fontsize':20})
```

```python
#for labeling the graph.
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('First Graph',fontdict={'fontname':'Comic Sans Ms','fontsize':20})
plt.show()
```

```python
#for legend of the graph.
plt.plot(x,y,'b*')
plt.legend(['single element'])
plt.show()
```

```python
plt.plot(x,y,'b*--')
plt.savefig('first graph.png',dpi  =200)
```

piechart.

```python
#for ploting piechart.
y= np.array([25,30,35,40])
name = ["Ajay","Vijay","Suresh","Sanjay"]

plt.pie(y,labels = name)
plt.show()
```

```python
#example.(how to show differently)

nameexplode = [0,0,0,0.2]
plt.pie(y,labels = name,explode =nameexplode)
plt.show()
```

```python
#shadow.
plt.pie(y,labels = name,explode = nameexplode,shadow = True)
plt.savefig('pie.png',dpi = 300)
plt.show()
```

```python
#giving color
mycolor = ['b','hotpink','black','m']
plt.pie(y,labels = name,explode =nameexplode,shadow = True,colors = mycolor)
plt.legend()
plt.show()
```

```python
#ploting boxplot graph which will be different.
test = np.random.normal(200,50,400)
plt.boxplot(test)
plt.show()
```

```python
#multiplle box plot.

test1 =np.random.normal(200,120,50)
test2 = np.random.normal(520,230,254)
test3 = np.random.normal(524,320,240)
data = [test1,test2,test3]
plt.boxplot(data)
plt.show()
```
