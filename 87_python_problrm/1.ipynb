```python
n= int(input("enter number: "))

count = 0
while (n>0):
    count = count +(n & 1)
    n= n//2
print(count)
```

### merge 2 sorted array in a single array.

```python
a1 = list(map(int,input("Enter 1st array. ").split()))
a2 = list(map(int,input("Enter 2nd array. ").split()))
a3 = []
i =j =0
while i <len(a1) and j<len(a2):
    if a1[i]>a2[j]:
        a3.append(a2[j])
        j +=1
    else:
        a3.append(a1[i])
        i +=1
while i<len(a1):
    a3.append(a1[i])
    i +=1
while j<len(a2):
    a3.append(a2[j])
    j +=1
print(a3)
```

### majority element i an array

```python
# brute force approch:
class Solution:
    def majority(self,arr:list[int])->int:
        n= len(arr)
        for i in range(0,n):
            count= 0
            for j in range(0,n):
                if arr[j] == arr[i]:
                    count +=1
            if count>n/2:
                return arr[i]
        return -1
obj = Solution()
arr = [2,3,4,5,3,3]
obj.majority(arr)
            
```

### Check Wether a number is stong or not!

```python
n = int(input("Enter. "))
sum1 = 0
org = n
while n !=0:
    i_d = n%10
    n= n//10
    fact = 1
    for i in range(1,i_d+1):
        fact *= i
    print(fact)
    sum1 +=fact
if sum1==org:
    print("number is strong. ")
else:
    print("number is not strong. ")
```
