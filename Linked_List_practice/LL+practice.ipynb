```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Create linked list: 1 → 2 → 3 → 4 → 5
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

def fun(head):
    if head is None:
        return
    if head.next and head.next.next is not None:
        print(head.data, " ", end="")
        fun(head.next)
    print(head.data, " ", end='')

fun(head)
print("\n")

# Q1. Replace max value with given value
def replace_max(value):
    temp = head
    max_node = temp
    while temp is not None:
        if temp.data > max_node.data:
            max_node = temp
        temp = temp.next
    max_node.data = value

replace_max(16)

# Q2. Sum of odd-positioned nodes (0-based index)
def sum_odd_nodes(head):
    temp = head
    counter = 0
    result = 0
    while temp is not None:
        if counter % 2 != 0:
            result += temp.data
        counter += 1
        temp = temp.next
    print("Sum of odd-positioned nodes:", result)

sum_odd_nodes(head)

# Final linked list print
def print_list(head):
    while head:
        print(head.data, end=" → ")
        head = head.next
    print("None")

print_list(head)

### REVERSE THE LINKED LIST--> Q3 of 2nd photo

def reverse(head):
    prev_node = None
    curr_node = head
    while curr_node !=None:
        next_node = curr_node.next
        curr_node.next = prev_node
        prev_node = curr_node
        curr_node = next_node
    return prev_node
head = reverse(head)
print("Reversed Linked List: ")
print_list(head)


## 4 of 2nd photo

```
